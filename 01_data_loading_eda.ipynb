- We scroll through each folder → each text file.
- We read the article content.
- We save the text and category.
- Finally, we build a DataFrame with two columns: text: article content and label: category.

```python
def load_bbc_dataset(path):
    texts = []
    labels = []
    for category in path.iterdir():
        if category.is_dir():
            for file in category.iterdir():
                if file.is_file():
                    with open(file, "r", encoding="latin-1") as f:
                        text = f.read()
                    texts.append(text)
                    labels.append(category.name)
    return pd.DataFrame({"text": texts, "label": labels})

```

- Let's run the upload function
- Display the first 5 articles to check everything.


```python
df = load_bbc_dataset(DATASET_DIR)
df.head()
```

The BBC dataset contains 2232 documents distributed across 5 news categories.

- We count how many documents there are for each category.
- We check whether the dataset is balanced (it is).

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="label", order=df["label"].value_counts().index)
plt.title("BBC Class Distribution")
plt.xlabel("Category")
plt.ylabel("Number of documents")
plt.show()
```

- Let's calculate the word count for each article.
- Let's look at the statistics: average, minimum, maximum...

```python
df["length"] = df["text"].apply(lambda x: len(x.split()))
df["length"].describe()
```

- Let's compare the length of articles in various categories.
- Useful for observing structural differences between documents.

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x="label", y="length")
plt.title("Document length by category")
plt.xlabel("Category")
plt.ylabel("Number of words")
plt.show()
```
